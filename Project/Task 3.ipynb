```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams['figure.figsize'] = (12,8)
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head(20)
```

```python
df.info()
```

```python
missing_cols = [col for col in df.columns if df[col].isnull().sum() > 0]
```

```python
df['2024 [YR2024]'].mean()
```

```python
cat_cols = df.select_dtypes(include = 'object')
num_cols = df.select_dtypes(exclude = 'object')
missing_cat_col = [col for col in missing_cols if col in cat_cols]
missing_num_col = [col for col in missing_cols if col in num_cols]
```

```python
for col in missing_cat_col:
    mode = df[col].mode()[0]
    df[col] = df[col].fillna(mode)
```

```python
for col in missing_num_col:
    mean = df[col].mean()
    df[col] = df[col].fillna(mean)
```

```python
world_data = df[df['Country Name'] == 'World']
world_population_growth = world_data[world_data['Series Name'] == 'Population growth (annual %)'].sum()[4:]
```

```python
world_population_growth
```

```python
years = [x[:4] for x in world_population_growth.index]
len(years)
```

```python
fig, axs = plt.subplots(5, figsize = (20,20))
fig.suptitle('Annual World Population Growth')
sns.barplot(ax = axs[0], x = years[:21], y=(world_population_growth.values)[:21])
sns.barplot(ax = axs[1], x = years[21:41], y=(world_population_growth.values)[21:41])
sns.barplot(ax = axs[2], x = years[41:61], y=(world_population_growth.values)[41:61])
sns.barplot(ax = axs[3], x = years[61:81], y=(world_population_growth.values)[61:81])
sns.barplot(ax = axs[4], x = years[81:], y=(world_population_growth.values)[81:])
```
